```python
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, Polygon
import folium
# import geoplot
# import geoplot.crs as gcrs
```

```python
df2019 = pd.DataFrame()
df2018 = pd.DataFrame()
df2019 = pd.read_csv("caracteristiques-2019.csv",sep=';')
#df2018 = pd.read_csv("caracteristiques-2018.csv",sep=',',encoding = "ISO-8859-1")
df2019['Date'] = df2019["jour"].astype(str) + "-" +df2019["mois"].astype(str)+ "-" + df2019["an"].astype(str)
df2019['Date'] = pd.to_datetime(df2019["Date"] + " " + df2019["hrmn"])
df2019 = df2019.drop(['jour','mois','an','hrmn','Num_Acc'],axis=1)
cols = df2019.columns.to_list()
newcol = cols[-1:]+cols[:-1]
df2019=df2019[newcol]
df2019 = df2019.loc[df2019["dep"].isin(["91","92","93","94","95","75","77","78"])].reset_index()
```

```python
df2019["adr"].value_counts().head(5)
```

```python
data = gpd.read_file("region-ile-de-france.geojson")
```

```python
df2019['long']=df2019['long'].replace(',','.', regex=True).apply(pd.to_numeric,errors='coerce')
df2019['lat']=df2019['lat'].replace(',','.', regex=True).apply(pd.to_numeric,errors='coerce')
```

```python

df2019.drop(df2019[df2019["long"]>10].index,inplace=True) #outlier
df2019=df2019.reset_index()
```

```python
df2019.size
```

```python
geometry = [Point(xy) for xy in zip(df2019['long'],df2019['lat'])]
```

```python
crs = {'init':'EPSG:4326'}
geo_df = gpd.GeoDataFrame(df2019, crs = crs, geometry = geometry)
```

```python
geo_df
```

```python
fig,ax = plt.subplots(figsize=(15,15))
data.plot(ax=ax,color='lightgrey')
geo_df.plot(ax=ax)
```
